Для системы заданной своей передаточной функцией: 
$$W(s)=\frac{k(Ts+1)}{s^2};~k = 10;~T = 2;$$
построить амплитудно-фазовую характеристику (АФХ).

1) Воспользуемся готовыми средствами (модуль control для Python)

```python
import matplotlib.pyplot, control, numpy

k = 10
T = 2
W = control.tf([T*k, k], [1, 0, 0])
print(W)
```

```python
matplotlib.pyplot.figure(figsize=(15, 10))
control.nyquist(W)
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('U(w)')
matplotlib.pyplot.ylabel('V(w)')
matplotlib.pyplot.show()
```

2) Построим более подробную АФХ, для того, чтобы выяснить как она себя ведёт в окрестности нуля.
$$W(i\omega) = -\frac{k(1 + iT\omega)}{\omega^2} = - \frac{k}{\omega^2} - i\frac{kT}{\omega} = U(\omega) + iV(\omega)$$

```python
def W(s):
    return k*(1 + T*s)/s**2
```

```python
a = numpy.array([W(1j*omega) for omega in numpy.arange(1, 500, 0.1)])

matplotlib.pyplot.figure(figsize=(15, 10))
matplotlib.pyplot.plot(a.real, a.imag, 'k')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('U(w)')
matplotlib.pyplot.ylabel('V(w)')

matplotlib.pyplot.show()
```
