Для системы с $n=5$, годограф Михайлова которой изображен на рисунке, найти распределение корней.

$D(0) \ne 0 \Longrightarrow \rho = 0$

Годограф дважды пересекает начало координат $\Longrightarrow \mu = 4$.

$l = 1;~ r = 0;~ l + r = l - r = 1$

Распишем $D(\lambda)$

$D(\lambda) = (\lambda - \lambda_1)(\lambda - \lambda_2)(\lambda - \lambda_3)(\lambda - \lambda_4)(\lambda - \lambda_5) = (\lambda - x_1 - iy_1)(\lambda^2 + y^2_2)(\lambda^2 + y^2_3)$

$D(i\omega) = (i\omega - x_1 - iy_1)(\omega^2 - y_2^2)(\omega^2 - y_3^2) = U(\omega) + iV(\omega)$

$U(\omega) = -x_1(\omega^2 - y_2^2)(\omega^2 - y_3^2)$

$V(\omega) = (\omega - y_1)(\omega^2 - y_2^2)(\omega^2 - y_3^2)$

$\Im{D(0)} = 0 \Longrightarrow y_1 = 0$

Рассмотрим пример с $\lambda_1 = x_1 = -1,~\lambda_{2,3} = \pm y_2 = \pm i,~\lambda_{4,5} = \pm y_2 = \pm 2i$

```python
import numpy, matplotlib.pyplot

def D(l):
    return (l + 1)*(l**2 + 1)*(l**2 + 4)
```

```python
a = numpy.array([D(1j*omega) for omega in numpy.arange(0, 2.25, 0.01)])

matplotlib.pyplot.figure(figsize=(15, 10))
matplotlib.pyplot.plot(a.real, a.imag, 'k')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('U(w)')
matplotlib.pyplot.ylabel('V(w)')

matplotlib.pyplot.show()
```
